```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

data = pd.read_csv("/content/ENB2012_data.csv")

plt.figure(figsize=(10, 8))
sns.heatmap(data.corr(), annot=True, cmap='YlOrRd')
plt.show()
```

```python
x  = data.iloc[:,:-2]
X  = x.values
y = data.iloc[:,-2:]
Y = y.values


from sklearn.model_selection import train_test_split
x_train , x_test , y_train , y_test = train_test_split(x , y , test_size=0.33 , random_state =0)
from sklearn.linear_model import LinearRegression
lin_reg1 = LinearRegression()

lin_reg1 = LinearRegression()
lin_reg1.fit(x_train,y_train)
y_pred_lin_reg = lin_reg1.predict(x_test)

from sklearn.metrics import r2_score
from sklearn.metrics import mean_squared_error

r2_score_lr = r2_score(y_pred_lin_reg,y_test)
print("Lineer Reg R2",r2_score_lr)
mse_lr = mean_squared_error(y_test, y_pred_lin_reg)
print("Mean Squared Linear",mse_lr)
```

```python
import matplotlib.pyplot as plt

# Plot for the first target variable
plt.figure(figsize=(8, 6))
plt.scatter(y_test.iloc[:, 0], y_pred_lin_reg[:, 0], color='blue')
plt.plot([y_test.iloc[:, 0].min(), y_test.iloc[:, 0].max()], [y_test.iloc[:, 0].min(), y_test.iloc[:, 0].max()], 'k--', lw=4)
plt.xlabel('True Values')
plt.ylabel('Predicted Values')
plt.title('Linear Regression - First Target Variable')
plt.grid(True)
plt.show()

# Plot for the second target variable
plt.figure(figsize=(8, 6))
plt.scatter(y_test.iloc[:, 1], y_pred_lin_reg[:, 1], color='red')
plt.plot([y_test.iloc[:, 1].min(), y_test.iloc[:, 1].max()], [y_test.iloc[:, 1].min(), y_test.iloc[:, 1].max()], 'k--', lw=4)
plt.xlabel('True Values')
plt.ylabel('Predicted Values')
plt.title('Linear Regression - Second Target Variable')
plt.grid(True)
plt.show()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsRegressor
from sklearn.metrics import r2_score, mean_squared_error

# Assuming the target variable is the last column of y
y = data.iloc[:, -1]  # Selecting the last column of y as the target variable for KNN

X = data.iloc[:, :-2].values  # Features
y = y.values  # Target variable

# Splitting the data into train and test sets
x_train, x_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=0)

# Instantiate and fit the KNN Regression model
knn_reg = KNeighborsRegressor(n_neighbors=5)  # You can adjust the number of neighbors as needed
knn_reg.fit(x_train, y_train)

# Predicting
y_pred_knn = knn_reg.predict(x_test)

# Evaluating
r2_score_knn = r2_score(y_test, y_pred_knn)
print("KNN Regression R2 Score:", r2_score_knn)

mse_knn = mean_squared_error(y_test, y_pred_knn)
print("Mean Squared Error:", mse_knn)
```

```python
import matplotlib.pyplot as plt

# Plotting
plt.figure(figsize=(10, 6))
plt.scatter(y_test, y_pred_knn, color='blue', label='Predicted vs. Actual')
plt.plot([min(y_test), max(y_test)], [min(y_test), max(y_test)], linestyle='--', color='red', label='Perfect Prediction')
plt.title('KNN Regression: Actual vs. Predicted')
plt.xlabel('Actual')
plt.ylabel('Predicted')
plt.legend()
plt.grid(True)
plt.show()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.svm import SVR
from sklearn.metrics import r2_score, mean_squared_error

# Assuming the target variables are the last two columns of y
y_data = data.iloc[:, -2:]  # Selecting the last two columns of y as the target variables

X = data.iloc[:, :-2].values  # Features
y_data = y_data.values  # Target variables

# Splitting the data into train and test sets
x_train, x_test, y_train, y_test = train_test_split(X, y_data, test_size=0.33, random_state=0)

# Selecting only one target variable (e.g., the last one)
y_train_single = y_train[:, -1]
y_test_single = y_test[:, -1]

# Instantiate and fit the Support Vector Machine (SVM) regressor
svm_reg = SVR()
svm_reg.fit(x_train, y_train_single)

# Predicting
y_pred_svm = svm_reg.predict(x_test)

# Evaluating
r2_score_svm = r2_score(y_test_single, y_pred_svm)
print("SVM R2 Score:", r2_score_svm)

mse_svm = mean_squared_error(y_test_single, y_pred_svm)
print("Mean Squared Error:", mse_svm)
```

```python
import matplotlib.pyplot as plt

# Plotting for the last target variable
plt.figure(figsize=(10, 6))
plt.scatter(y_test_single, y_pred_svm, color='blue', label='Predicted vs. Actual')
plt.plot([min(y_test_single), max(y_test_single)], [min(y_test_single), max(y_test_single)], linestyle='--', color='red', label='Perfect Prediction')
plt.title('SVM Regression: Actual vs. Predicted (Last Target Variable)')
plt.xlabel('Actual')
plt.ylabel('Predicted')
plt.legend()
plt.grid(True)
plt.show()
```
